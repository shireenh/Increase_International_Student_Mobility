# Read Predictors data

```python
predictors_df = pd.read_csv('data/Primary_Commodity_Price_System_PCPS.csv', skiprows=1)
```

```python
predictors_df.head()
```

```python
predictors_df['year'] = predictors_df.Date.apply(lambda x: int(x.split(' ')[1]))
predictors_df.head()
```

```python
predictors_yearly_df = predictors_df.groupby('year').mean()
cols = predictors_yearly_df.columns
predictors_yearly_df = predictors_yearly_df.join(predictors_yearly_df[cols].shift(), rsuffix='_1y_shift')
predictors_yearly_df = predictors_yearly_df.join(predictors_yearly_df[cols].shift(2), rsuffix='_2y_shift')
predictors_yearly_df = predictors_yearly_df.loc[2003:2018]
```

```python
predictors_yearly_df.head()
```

# Read response data (Sponsored Students Counts)

```python
response_df = pd.read_csv("data/sponsor_data_countrywise_backup.csv").set_index('country')
response_df.columns.name = 'year'
response_df = response_df.stack()
response_df.name = 'students_count'
response_df = response_df.to_frame().reset_index()
response_df['year'] = response_df.year.astype(int)
response_df.to_csv('data/sponsored_students_counts.csv', index=False)
```

```python
mask = response_df.country.apply(lambda x: re.search('.*Total.*', x) is None)
response_df = response_df[mask]
```

```python
years_df = response_df.groupby('year').students_count.sum().sort_index()

traces = []
traces.append(go.Bar(
        x=years_df.index.values,
        y=years_df.values,
        name='Sponsored Student Count',
        opacity = 0.5))


layout = dict(
    title = 'Total Number of Sponsored Students By Year',
    xaxis = dict(title = 'Year'),
    
    yaxis = dict(
        title = 'Number of Sponsored Students',
        #range = [0, 250]
    ),
    
    legend=dict(
        x=0.3,
        y=1.1,
        traceorder='normal',
        font=dict(
            family='sans-serif',
            size=12,
            color='#000'
        ),
        orientation="h"
    )
,
)

fig = dict(data=traces, layout=layout)
iplot(fig)
```

```python
top_countries_df = response_df.groupby('country').students_count.sum().sort_values(ascending=False).head(30)
top_countries = top_countries_df.index.values
```

```python
traces = []
traces.append(go.Bar(
        x=top_countries_df.index.values,
        y=top_countries_df.values,
        name='Sponsored Student Count',
        opacity = 0.5))


layout = dict(
    title = 'Total Number of Sponsored Students By Top 30 Countries',
    xaxis = dict(title = 'Year'),
    
    yaxis = dict(
        title = 'Number of Sponsored Students',
        #range = [0, 250]
    ),
    
    legend=dict(
        x=0.3,
        y=1.1,
        traceorder='normal',
        font=dict(
            family='sans-serif',
            size=12,
            color='#000'
        ),
        orientation="h"
    )
,
)

fig = dict(data=traces, layout=layout)
iplot(fig)
```

# Visualise

## Correlation

```python
from scipy.stats import pearsonr

df = response_df.set_index(['country', 'year']).unstack('country')
```

```python
df.columns = df.columns.droplevel(0)
```

```python
df = df.loc[:2018][top_countries]
df.columns
```

```python
CPI_Country_df = predictors_yearly_df.join(df)
```

```python
corr_pvalues_df = pd.DataFrame(np.nan, index=CPI_Country_df.columns[30:60],columns=CPI_Country_df.columns[0:30])

for i in range(0,30): 
    index = CPI_Country_df.columns[i] #str: CPI name
    CPI_corr_pvalue = {} #dic: corr for one CPI with each country
    for j in range(30,60):
        country = CPI_Country_df.columns[j] #str: country name
        c_p = pearsonr(CPI_Country_df.iloc[:,j],CPI_Country_df.iloc[:,i]) # (corr, pvalue)
        CPI_corr_pvalue[country] = c_p
    corr_pvalues_df[index] = pd.Series(CPI_corr_pvalue)
    
```

```python
corr_pvalues_df
```

## Heatmap

```python
inve_country_corr_pvalues_df = corr_pvalues_df.loc[corr_pvalues_df.index.values[::-1]]
```

```python
inve_bin_pvalues = [] #list of list of float
for i, cnty in enumerate(inve_country_corr_pvalues_df.values):
    cnty_pvalues = []
    for j, idx in enumerate(cnty):
        pvalue = inve_country_corr_pvalues_df.values[i][j][1]
        if pvalue <= 0.05:
            cnty_pvalues.append(1) #reject H0 and the corr is significant
        else:
            cnty_pvalues.append(0) #accept H0 and the corr is insignificant
    inve_bin_pvalues.append(cnty_pvalues)
    
```

```python
inve_corrs = [] #list of list of float
for i, cnty in enumerate(inve_country_corr_pvalues_df.values):
    cnty_corrs = []
    for j, idx in enumerate(cnty):
        corr = inve_country_corr_pvalues_df.values[i][j][0]
        cnty_corrs.append(corr) 

    inve_corrs.append(cnty_corrs)
    
```

```python
trace_corr = go.Heatmap(z = inve_corrs,
                        x = inve_country_corr_pvalues_df.columns.values,
                        y = inve_country_corr_pvalues_df.index.values,
                        )
layout_corr = go.Layout(
    title="Correlation Between the Number of Sponsored Students and Commodity Indices",
    autosize=False,
    width=900,
    height=900)

iplot(go.Figure(data=[trace_corr], layout=layout_corr))
```

```python
inve_corrs_sig = [] #list of list of float
for i, cnty in enumerate(inve_country_corr_pvalues_df.values):
    inve_corrs_sig_cnty = []
    for j, idx in enumerate(cnty):
        corr = inve_country_corr_pvalues_df.values[i][j][0]
        pvalue = inve_country_corr_pvalues_df.values[i][j][1]
        if pvalue <= 0.05: #reject H0, the corr is sig
            inve_corrs_sig_cnty.append(corr)
        else:
            inve_corrs_sig_cnty.append(np.nan)
            

    inve_corrs_sig.append(inve_corrs_sig_cnty)
```

```python
trace_corr_sig = go.Heatmap(z = inve_corrs_sig,
                            x = inve_country_corr_pvalues_df.columns.values,
                            y = inve_country_corr_pvalues_df.index.values,
                            )
layout_corr_sig = go.Layout(
    title="Significant Correlation Between the Number of Sponsored Students and Commodity Indices",
    autosize=False,
    width=900,
    height=900)

iplot(go.Figure(data=[trace_corr_sig], layout=layout_corr_sig)) #transparent grid means insignificant correlation
```
